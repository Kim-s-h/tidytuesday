```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rc('font', size=12) #controls default text sizes
plt.rc('axes', titlesize=14) #fontsize of the axes title
plt.rc('axes', labelsize=12) #fontsize of the x and y labels
plt.rc('xtick', labelsize=12) #fontsize of the tick labels
plt.rc('ytick', labelsize=12) #fontsize of the tick labels
plt.rc('legend', fontsize=12) #legend fontsize
plt.rc('figure', titlesize=14) #fontsize of the figure title
plt.rcParams['figure.dpi'] = 300 #high resolution
sns.set_style('whitegrid')
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('NetflixViewingHistory.csv', parse_dates=['Date'])
```

```python
data['Title'], data['Season'], data['Episodes'] =  data['Title'].str.split(':',2).str
```

```python
data
```

```python
mask = (data['Date'] >= '2017-01-01') & (data['Date'] <= '2020-12-31')
```

```python
data2 = data.loc[mask]
```

```python
tv_show = data2[data2['Season'].notnull()]
```

```python
grouped_title = tv_show.groupby(['Title'])
```

```python
grouped_title.agg('count').sort_values(by=['Date'], ascending=False)
```

```python
ordered_tv_show = grouped_title.agg('count').sort_values(by=['Date'], ascending=False)
```

```python
ordered_tv_show.reset_index(inplace=True)
```

```python
most_watched = ordered_tv_show['Title'][0:10].tolist()
```

```python
tv_show_most_watched = tv_show[tv_show['Title'].isin(most_watched)]
```

```python
tv_show_most_watched['Year'] = tv_show_most_watched['Date'].dt.year
```

```python
tv_show_m_grouped = tv_show_most_watched.groupby(['Title','Year'])
```

```python
final_data = tv_show_m_grouped.agg('count')
final_data.reset_index(inplace=True)
```

```python
final_data = final_data[['Title','Year','Date']]
```

```python
final_data_wide = final_data.pivot(index="Title", columns = "Year", values = "Date")
```

```python
plt.figure(figsize=(16,8))
final_data_wide.plot.barh(stacked = True, alpha = 0.7)
plt.savefig('Netflix_Python_stacked.png', dpi = 400, bbox_inches='tight')
```

```python
sns.barplot(x ='Date', y = "Title", data=final_data, 
            orient = 'h', hue="Year")
```

```python
plt.figure(figsize=(14,10))
g=sns.FacetGrid(final_data, col = "Year", col_wrap=2, hue="Year")
g.map(sns.barplot, "Date", "Title", alpha = 0.7)
g.set_axis_labels("", "")
plt.savefig('Netflix_Python_byYear.png', dpi = 300)
```
